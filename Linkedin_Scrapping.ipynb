```python
informations=[]
for i in range(15):
    url = 'https://www.linkedin.com/jobs/search?keywords=Data%20scientist&location=Maroc&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum='+str(i*25)
    response = requests.get(url)
    soup = BeautifulSoup(response.content,'html.parser')
    listes=soup.find_all("div", class_="base-card relative w-full hover:no-underline focus:no-underline base-card--link base-search-card base-search-card--link job-search-card") # ceci dépend du site source (c’est le cas de l’urlé)
    
    for liste in listes :

        job_title= liste.find('h3', class_='base-search-card__title').text.replace('\n', '')
        company= liste.find('h4', class_='base-search-card__subtitle').text.replace('\n', '')
        location= liste.find('span', class_="job-search-card__location").text.replace('\n', '')
        url = liste.find('a', class_="base-card__full-link",href=True)["href"].replace('\n', '').strip()
        response1 = requests.get(url)
        soup1 = BeautifulSoup(response1.content,'html.parser')
        description=soup1.find("div", class_="show-more-less-html__markup")
        if description is not None :

            information =[job_title.strip(), company.strip(), location.strip(), url,description.text.replace('\n', '')]
            informations.append(information)



    
```
